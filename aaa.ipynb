```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
import csv
from scipy import signal

hdul_x = open('x_data.txt')
hdul_y = open('y_data_quad.txt')
array_x0 = hdul_x.readlines()
array_y0 = hdul_y.readlines()
array_x = [float(numeric_string) for numeric_string in array_x0]
array_y = [float(numeric_string) for numeric_string in array_y0]
a1 = 0
a2 = 0
for i in range(1, len(array_x)):
    a1 += (array_y[i] * pow(array_x[i], 2))
    a2 += pow(array_x[i], 4)
a = a1 / a2
print(a)
def qua(x,a):
    return a*pow(x,2)
constants = curve_fit(qua, array_x, array_y)
a_fit = constants[0]
print(a_fit)
fit = []
for i in array_x:
    fit.append(qua(i, a_fit))
plt.plot(array_x, array_y)
plt.plot(array_x, fit)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('trying to find a')
plt.show()
```

```python
hdul_y_nl = open('y_data_nonlinear.txt')
array_y_nl0 = hdul_y_nl.readlines()
array_y_nl = [float(numeric_string) for numeric_string in array_y_nl0]
def function(x, a, b, c):
    return a * np.exp(-b * np.sin(x)) + c * x
constants = curve_fit(function, array_x, array_y_nl)
a_fit = constants[0][0]
b_fit = constants[0][1]
c_fit = constants[0][2]
print(a_fit, b_fit, c_fit)
plt.xlabel('x')
plt.ylabel('y')
plt.title('trying to find a, b and c')
plt.plot(array_x, array_y_nl)
plt.grid()
plt.show()
```

```python
hdul_x1 = open('x_data1.txt')
hdul_y1 = open('y_data1.txt')
array_x10 = hdul_x1.readlines()
array_y10 = hdul_y1.readlines()
array_x1 = [float(numeric_string) for numeric_string in array_x10]
array_y1 = [float(numeric_string) for numeric_string in array_y10]

def gauss(x, a, mu, sig):
    return a*np.exp(-(x-mu)**2/(2*sig**2))

constants = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 45, 0), (47, 47, 40)))
constants1 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 200, 0), (100, 205, 30)))
constants2 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 498, 0), (250, 501, 20)))
constants3 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 598, 0), (270, 601, 20)))
constants4 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 698, 0), (250, 701, 45)))
constants5 = curve_fit(gauss, array_x1, array_y1, bounds = ((0,990,0), (70, 995, 300)))
constants6 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 1198, 0), (130, 1201, 30)))
constants7 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 1498, 0), (350,1501, 85)))
constants8 = curve_fit(gauss, array_x1, array_y1, bounds = ((0, 1998, 0), (170, 2001, 280)))

def fun(x):
    return gauss(x, constants[0][0], constants[0][1], constants[0][2]) + gauss(x, constants1[0][0], constants1[0][1], constants1[0][2]) + gauss(x, constants2[0][0], constants2[0][1], constants2[0][2]) + gauss(x, constants3[0][0], constants3[0][1], constants3[0][2]) + gauss(x, constants4[0][0], constants4[0][1], constants4[0][2]) + gauss(x, constants5[0][0], constants5[0][1], constants5[0][2]) + gauss(x, constants6[0][0], constants6[0][1], constants6[0][2]) + gauss(x, constants7[0][0], constants7[0][1], constants7[0][2]) + gauss(x, constants8[0][0], constants8[0][1], constants8[0][2])

print(constants[0][0], constants[0][1], constants[0][2])
print(constants1[0][0], constants1[0][1], constants1[0][2])
print(constants2[0][0], constants2[0][1], constants2[0][2])
print(constants3[0][0], constants3[0][1], constants3[0][2])
print(constants4[0][0], constants4[0][1], constants4[0][2])
print(constants5[0][0], constants5[0][1], constants5[0][2])
print(constants6[0][0], constants6[0][1], constants6[0][2])
print(constants7[0][0], constants7[0][1], constants7[0][2])
print(constants8[0][0], constants8[0][1], constants8[0][2])

plt.plot(array_x1, array_y1)
plt.plot(fun(range(0, 2500)), 'r')
plt.xticks(color='white')
plt.yticks(color='white')
plt.grid()
plt.xlabel('x', color='white')
plt.ylabel('y', color='white')
plt.show()
```

```python
file = open('J0854+2006.csv')
csvreader = csv.reader(file)
header = next(csvreader)
rows00 = []
for row in csvreader:
    rows00.append(row)

rows0 = []
for i in rows00:
    rows0 += i
rows = [float(numeric_string) for numeric_string in rows0]
mjd = []
for i in range(0,212):
    mjd.append(rows[3*i])
flux = []
for i in range(0,212):
    flux.append(rows[3*i+1])
flux_err = []
for i in range(0,212):
    flux_err.append(rows[3*i+2])

def s_func(x, Sm, xm, tau):
    return Sm*np.exp(-abs(x-xm)/tau)

#constants = curve_fit(s_func, mjd, flux, bounds = ((0, 54499, 0), (2.3, 54501, 100)))
constants1 = curve_fit(s_func, mjd, flux, bounds = ((0,54627,0), (3.1, 54629, 250)))
#constants2 = curve_fit(s_func, mjd, flux, bounds = ((0, 54705, 0), (0.6, 54715, 40)))
constants3 = curve_fit(s_func, mjd, flux, bounds = ((0, 54913, 0), (3.2,55915, 110)))
#constants4 = curve_fit(s_func, mjd, flux, bounds = ((0,55075, 0), (2.5, 55085, 150)))
constants5 = curve_fit(s_func, mjd, flux, bounds = ((0, 55197, 0), (11, 55199, 110)))

def fun(x):
    return s_func(x, constants1[0][0], constants1[0][1], constants1[0][2]) + s_func(x, constants3[0][0], constants3[0][1], constants3[0][2]) + s_func(x, constants5[0][0], constants5[0][1], constants5[0][2])

print(constants1[0][0], constants1[0][1], constants1[0][2])
print(constants3[0][0], constants3[0][1], constants3[0][2])
print(constants5[0][0], constants5[0][1], constants5[0][2])

def doppler(Sm, tau):
    return (((1.548*(10**-32))*(Sm*((5.071*9.461)**2)*(10**48))/(225*(tau**2)*1.306))/(5*10**10))**(1/3)

print('\n', doppler(constants5[0][0], constants5[0][2]))

plt.figure(figsize=(35,25))
#plt.scatter(mjd, flux)
plt.errorbar(mjd, flux, yerr=flux_err, ecolor='black')
plt.plot(mjd, fun(mjd), 'r')
plt.xticks(color='white', fontsize=30)
plt.yticks(color='white', fontsize=30)
plt.grid()
plt.title('J0854+2006', color='white', fontsize=40)
plt.xlabel('MJD', color='white', fontsize=30)
plt.ylabel('Flux density (Jy)', color='white', fontsize=30)
plt.show()
```
